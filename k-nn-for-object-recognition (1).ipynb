```python
import tensorflow as tf

cifar_10 = tf.keras.datasets.cifar10

(train_image , train_labels) , (test_image , test_labels) = cifar_10.load_data()
```

```python
from scipy import stats
class NN:
    
    def __inti__(self):
        pass
    
    def train_model(self,train_image,train_labels):
        self.train_image = train_image
        self.train_labels = train_labels
    
    def distance_metric(self,r,n):
        ans = r
        dimension = len(ans.shape)
        for i in range(n-1):
            ans *= r
        for i in range(dimension-1):
            ans = np.sum(ans,axis=1)
        return ans**(1/n)
        
    def predict(self,test_image):
        n = test_image.shape[0]
        test_labels = np.zeros(n , dtype = np.int32)
        
        for i in range(n):
            distance = np.abs(self.train_image - test_image[i:i+1,:])
            distance = self.distance_metric(distance,2)
            idx = np.argmin(distance)
            test_labels[i] = self.train_labels[idx]
            
        return test_labels
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets

mnist = datasets.load_digits()

X = mnist.data
y = mnist.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
knn = OneVsRestClassifier(KNeighborsClassifier(n_neighbors = 1))
knn.fit(X_train , y_train)
prediction = knn.predict(X_test)
ar = prediction - y_test
print(np.where(ar == 0)[0].shape[0])
```

```python
knn = NN()
knn.train_model(X_train , y_train)
prediction = knn.predict(X_test)
ar = prediction - y_test
print(np.where(ar == 0)[0].shape[0])
```
